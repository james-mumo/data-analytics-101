```python
import matplotlib.pyplot as plt

ages = [22, 25, 27, 30, 30, 31, 32, 33, 34, 35, 36, 38, 40, 42, 45, 50]
plt.hist(ages, bins=5, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# box plot
```

```python
import matplotlib.pyplot as plt

scores = [70, 75, 80, 80, 85, 85, 85, 90, 95]

plt.plot(scores, [1]*len(scores), 'bo', ms=10)  # 'bo' for blue dots, ms for marker size
plt.title('Dot Plot of Exam Scores')
plt.xlabel('Scores')
plt.yticks([])  # Hides the y-axis ticks
plt.show()
```

```python
# scatter plot
```

```python
import matplotlib.pyplot as plt

heights = [150, 160, 165, 170, 175, 180, 185]
weights = [50, 55, 60, 65, 70, 75, 80]

plt.scatter(heights, weights, color='red')
plt.title('Height vs Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.show()
```
